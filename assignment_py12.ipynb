## Schelling model desciption <a id='1'></a>

The Schelling model is an abstract model that explores the effect of individual relocation behaviours on macroscale segregation.
The Schelling Agent Based Moodel (ABM) includes: a space, and agents and objects that temporally interact with each other and with the model space.
We describe each component of the Schelling ABM hereafter.

### Space
The space is represented as a grid of square cells. Each grid cell represents a spatial unit that one agent can occupy.
Each spatial unit has 8 immediate surrounding neighbours defined as the "Moore neighbourhood".

### Agents
The model includes one type of agent representing an individual or a household.
Each agent has the following state parameters (i.e., variables that dictate the state of the agent during the simulation):
1. `happy?`           ; indicates whether the agent is happy or not (happy agents have a percentage of surrounding similar agents above the `similarity_threshold`)
2. `similar-nearby`   ; the number of neighboring spatial units that have an agent similar to the agent under analysis
3. `other-nearby`     ; the number of neighbouring agents that are different from the agent under analysis
4. `total-nearby`     ; the total numebr of neighbouring agents

Agents aim to settle in a spatial unit that makes them happy by being surrounding with a high ratio of simiar agents in their Moore neighbourhood. If the agent is unhappy (due to being surrounded by a high ratio of agents different from itself), the agent moves from the current spatial unit to another one randomly. The decision tree of the agents can be summarised as follows:
- if `similar-nearby / total-nearby > similarity_threshold` (i.e., surrrounded by many similar agents) --> happy --> remain in current spatial unit
- if `other-nearby / total-nearby > similarity_threshold` (i.e., surrrounded by many different agents) --> unhappy --> move to another random spatial unit

### Temporality
Each time step in the model represents an iteration where each agent checks whether they are happy or not. Following the previously stated decision tree, each agent decides whether it will move or not.

Part 1: What are ABMs, and what are their key features? Cite at least 2 examples.

**Answer:** Agent Based Models are a type of simulation in which individuals (agents) interact with each other in some virtual space (grids or networks). The individuals are assigned some characteristics and rules, they interact with each other based on these rules. Iterating the model over time allows a larger system to be understood from micro-level interactions. Therefore, changing the agent's characteristics and rules will change outcome of the larger system's behaviour.

Some examples of ABMs applications include:

Jenkins, K., Surminski, S., Hall, J. and Crick, F. (2017). Assessing surface water flood risk and management strategies under future climate change: Insights from an Agent-Based Model. Science of the Total Environment, 595, pp.159-168. Where the authors attempt to understand the interactions between homeowners, insurers, local governments, property developers, and banks to understand the dynamics between different surface water flooding adaptation options, and how properties can become more resilient due to action from homeowners, governments, and insurers.

Lux, T. (1998). The socio-economic dynamics of speculative markets: interacting agents, chaos, and the fat tails of return distributions. Journal of Economic Behavior & Organization, 33(2), pp.143-165. In this example, an ABM is applied in a socio-economic context in which agents (traders) interacted with eachother to simulate financial markets in order to understand how returns changed due to the different behaviours the agents exhibit.

It is clear that ABMs have a wide range of applications and offer deep insight on how small-scale (and often random) interactions between agents and interactions between agents and the environment form complex and collective patterns.

Part 3: Critically Evaluate the Schelling model (is it fit for purpose?)

**Answer:** Despite the simplicity of the Schelling Model, it is a powerful tool to understand how segregation can occur in a population without any one agent deliberately acting in a prejudiced way, this is the key nuance the Schelling Model offers. 

However, it does not offer an explanation to *why* segregation occurs. To answer this question, the model must include information about the environment, i.e. neighbourhood boundaries, access to key local services and opportunities, etc. Furthermore, apart from colour, all agents are identical. In reality, people are unique, some people can move freely whereas others can't, some groups are more tolerant to other people, but some are not. Moreover, the model must include external forces that may "push" segregation, for example local governments that enforce zoning laws, or socio-economic inequalities that impact where people can move to. 

Therefore, the Schelling model is a useful tool to understand the dynamics of segregation and to explore how segregation changes due to different scenarios, but not to predict where segregation occurs. Also, without further development, the model does not provide insight on the causal factors that drive segregation.
